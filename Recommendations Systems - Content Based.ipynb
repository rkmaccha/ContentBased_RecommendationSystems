```python
# movies["movie_id"] - No issues - the column consists a int dtype with the id of the movie
# movies["title"] - No issues - the column consists a object (string) dtype with the title of the movie
# movies["genres"] - the column consists a object (string) dtype with the genres of the movie as a string [list of dictionary]
# movies["genres"] - Need to convert the value as a list of genres
```

```python
movies["genres"].iloc[0]
```

```python
# ast Module -  expands to Abstract Syntax Tree. The ast module helps Python applications to process trees of the Python abstract syntax grammar.

# The literal_eval safely evaluate an expression node or a string containing a Python literal or container display. 
#The string or node (a file) provided may only consist of the following Python literal structures: strings, bytes, numbers, tuples, lists, dicts, sets, booleans, and None.
# movies["genres"] data is a string with a list of dictionaries, we will use ast.literal_eval to convert into list of dictionaries
import ast
```

```python
def Convert(obj):
    tempL=[]
    for i in ast.literal_eval(obj):
        tempL.append(i['name'])
    return tempL
```

```python
movies["genres"] = movies["genres"].apply(Convert)
```

```python
movies["genres"].head()
```

```python
type(movies["genres"][0])
```

```python
movies['keywords'] = movies['keywords'].apply(Convert)
```

```python
movies['keywords'].head()
```

```python
type(movies['keywords'][0])
```

```python
def GetCastList(obj):
    tempL=[]
    counter =0
    for i in ast.literal_eval(obj):
        if counter <3 :
            tempL.append(i['name'])
            counter+=1
        else:
            break
    return tempL
```

```python
movies['cast'] = movies['cast'].apply(GetCastList)
```

```python
type(movies['cast'][0])
```

```python
movies['crew'][0]
```

```python
def GetCrewList(obj):
    tempL=[]
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            tempL.append(i['name'])
            break
    return tempL
```

```python
movies['crew'] = movies['crew'].apply(GetCrewList)
```

### TEXT VECTORISATION

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies_final['tags']).toarray()
```

```python
vectors[0] # First movie in the list Avatar's vector information
```

```python
cv.get_feature_names() # total max_features words overall 
```

```python
len(cv.get_feature_names())
```

```python
import nltk 
```

```python
from nltk.stem.porter import PorterStemmer
PS = PorterStemmer()
```

```python
def stem(text):
    tempL=[]
    for i in text.split():
        tempL.append(PS.stem(i))
    return " ".join(tempL)
```

```python
stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCameron')
```

```python
movies_final['tags'] = movies_final['tags'].apply(stem)
```

```python
vectors = cv.fit_transform(movies_final['tags']).toarray()
cv.get_feature_names()
```

### Cosine Distane  - Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0] # Similarity of the First Movie with other movies in the data
```

```python
def Recommend(movie):
    movie_index = movies_final[movies_final['title']==movie].index[0]
    distances= similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x : x[1])[1:6]
    
    for movie in movies_list:
        #print(movie)
        #print(movie[0])
        print(movies_final.iloc[movie[0]].title)
```

```python
Recommend("Avatar") # print(movie) - prints the index and the similarity
```
